### CHI-MERGE

```python
import numpy as np
#Chi-Merge 
def ChiMerge(df,variable,flag,confidenceVal=3.841,bin=10,sample=None):  
    '''
    param df:DataFrame
    param variable:str
    param flag:str    
    param confidenceVal:float
    param bin：int           
    param sample: int         
    note: 
    return :DataFrame
    '''    
    import pandas as pd
    import numpy as np
    
    
    #whether sampling
    if sample != None:
        df = df.sample(n=sample)
    else:
        df   
        
    #import dataset 
    total_num = pd.DataFrame({'total_num': total_num})  #create a dataframe to store the original output
    positive_class = df.groupby([variable])[flag].sum()  #the number of positive sample which need to be binning
    positive_class = pd.DataFrame({'positive_class': positive_class})  #create a new df to store the postive output
    regroup = pd.merge(total_num, positive_class, left_index=True, right_index=True,
                       how='inner')  # combin total_num and positive_class
    regroup.reset_index(inplace=True)
    regroup['negative_class'] = regroup['total_num'] - regroup['positive_class']  #the number of negative sample which need to be binning
    regroup = regroup.drop('total_num', axis=1)
    np_regroup = np.array(regroup)  #transform dataframe to numpy（to increase working efficiency）
    #print
    print('*'*20)
    print(variable)
    #Process intervals without positive or negative samples and merge the intervals (to avoid errors in the calculation of the chi-square value)
    if sum(regroup['negative_class'])!=0 and sum(regroup['positive_class'])!=0:
        i = 0
        while (i <= np_regroup.shape[0] - 2):
            if ((np_regroup[i, 1] == 0 and np_regroup[i + 1, 1] == 0) or ( np_regroup[i, 2] == 0 and np_regroup[i + 1, 2] == 0)):
                np_regroup[i, 1] = np_regroup[i, 1] + np_regroup[i + 1, 1]  # positive sample
                np_regroup[i, 2] = np_regroup[i, 2] + np_regroup[i + 1, 2]  # negative sample
                np_regroup[i, 0] = np_regroup[i + 1, 0]
                np_regroup = np.delete(np_regroup, i + 1, 0)
                i = i - 1
            i = i + 1

        #Chi-square calculation of two adjacent intervals
        chi_table = np.array([])  # Create an array storing the chi-square values of two adjacent intervals
        for i in np.arange(np_regroup.shape[0] - 1):
            chi = (np_regroup[i, 1] * np_regroup[i + 1, 2] - np_regroup[i, 2] * np_regroup[i + 1, 1]) ** 2 \
              * (np_regroup[i, 1] + np_regroup[i, 2] + np_regroup[i + 1, 1] + np_regroup[i + 1, 2]) / \
              ((np_regroup[i, 1] + np_regroup[i, 2]) * (np_regroup[i + 1, 1] + np_regroup[i + 1, 2]) * (
              np_regroup[i, 1] + np_regroup[i + 1, 1]) * (np_regroup[i, 2] + np_regroup[i + 1, 2]))
            chi_table = np.append(chi_table, chi)
        #print

        #Combine the two intervals with the smallest chi-square value (the core of chi-Merge)
        while (1):
            if (len(chi_table) <= (bin - 1) and min(chi_table) >= confidenceVal):
                break
            chi_min_index = np.argwhere(chi_table == min(chi_table))[0] 
            np_regroup[chi_min_index, 1] = np_regroup[chi_min_index, 1] + np_regroup[chi_min_index + 1, 1]
            np_regroup[chi_min_index, 2] = np_regroup[chi_min_index, 2] + np_regroup[chi_min_index + 1, 2]
            np_regroup[chi_min_index, 0] = np_regroup[chi_min_index + 1, 0]
            np_regroup = np.delete(np_regroup, chi_min_index + 1, 0)

            if (chi_min_index == np_regroup.shape[0] - 1):  # When the minimum value is tested in the last two intervals
                # Calculate the chi-square value of the current interval and the previous interval after merging and replace
                chi_table[chi_min_index - 1] = (np_regroup[chi_min_index - 1, 1] * np_regroup[chi_min_index, 2] - np_regroup[chi_min_index - 1, 2] * np_regroup[chi_min_index, 1]) ** 2 \
                                               * (np_regroup[chi_min_index - 1, 1] + np_regroup[chi_min_index - 1, 2] + np_regroup[chi_min_index, 1] + np_regroup[chi_min_index, 2]) / \
                                           ((np_regroup[chi_min_index - 1, 1] + np_regroup[chi_min_index - 1, 2]) * (np_regroup[chi_min_index, 1] + np_regroup[chi_min_index, 2]) * (np_regroup[chi_min_index - 1, 1] + np_regroup[chi_min_index, 1]) * (np_regroup[chi_min_index - 1, 2] + np_regroup[chi_min_index, 2]))
                # Delete the chi-square value of previous interval
                chi_table = np.delete(chi_table, chi_min_index, axis=0)

            else:
                # Calculate the chi-square value of the current interval and the previous interval after merging and replace
                chi_table[chi_min_index - 1] = (np_regroup[chi_min_index - 1, 1] * np_regroup[chi_min_index, 2] - np_regroup[chi_min_index - 1, 2] * np_regroup[chi_min_index, 1]) ** 2 \
                                           * (np_regroup[chi_min_index - 1, 1] + np_regroup[chi_min_index - 1, 2] + np_regroup[chi_min_index, 1] + np_regroup[chi_min_index, 2]) / \
                                           ((np_regroup[chi_min_index - 1, 1] + np_regroup[chi_min_index - 1, 2]) * (np_regroup[chi_min_index, 1] + np_regroup[chi_min_index, 2]) * (np_regroup[chi_min_index - 1, 1] + np_regroup[chi_min_index, 1]) * (np_regroup[chi_min_index - 1, 2] + np_regroup[chi_min_index, 2]))
                # Calculate the chi-square value of the merged current interval and the next interval and replace
                chi_table[chi_min_index] = (np_regroup[chi_min_index, 1] * np_regroup[chi_min_index + 1, 2] - np_regroup[chi_min_index, 2] * np_regroup[chi_min_index + 1, 1]) ** 2 \
                                           * (np_regroup[chi_min_index, 1] + np_regroup[chi_min_index, 2] + np_regroup[chi_min_index + 1, 1] + np_regroup[chi_min_index + 1, 2]) / \
                                       ((np_regroup[chi_min_index, 1] + np_regroup[chi_min_index, 2]) * (np_regroup[chi_min_index + 1, 1] + np_regroup[chi_min_index + 1, 2]) * (np_regroup[chi_min_index, 1] + np_regroup[chi_min_index + 1, 1]) * (np_regroup[chi_min_index, 2] + np_regroup[chi_min_index + 1, 2]))
                # Delete chi-squared value before replacement
                chi_table = np.delete(chi_table, chi_min_index + 1, axis=0)
        #print()

        #Save the results as a new data frame
        result_data = pd.DataFrame()  
        result_data['variable'] = [variable] * np_regroup.shape[0]  # the first column of result：names of feature 
        list_temp = []
        for i in np.arange(np_regroup.shape[0]):
            if i == 0:
                x = str(np_regroup[i, 0])
            elif i == np_regroup.shape[0] - 1:
                x = str(np_regroup[i - 1, 0]) + '+'
            else:
                x = str(np_regroup[i - 1, 0]) + ',' + str(np_regroup[i, 0])
            list_temp.append(x)
        result_data['interval'] = list_temp  #the second column：interval
        result_data['flag_0'] = np_regroup[:, 2]  # the third column：numbers of negative sample
        result_data['flag_1'] = np_regroup[:, 1]  # the forth column：numbers of postive sample

    return result_data
```

#### 计算IV值

```python
###calculate the value of IV
def discrete_iv(df,col_name):
    import numpy as np
    import pandas as pd
    import math
    tmp4_1 = df[[col_name,'label']][df[col_name]!=-99]
    list1 = list(tmp4_1.groupby(col_name).size().index)
    N = len(list1)

    result_data = pd.DataFrame()
    result_data['variable'] = [col_name]*N
    result_data['interval'] = list1
    #numbers of negative sample
    result_data['interval_neg_cnt'] = np.array(tmp4_1.groupby([col_name])['label'].count())-np.array(tmp4_1.groupby([col_name])['label'].sum())
    result_data['interval_neg_cnt'] = result_data['interval_neg_cnt'].fillna(0)

    #numbers of postive sample
    result_data['interval_pos_cnt'] = np.array(tmp4_1.groupby([col_name])['label'].sum())
    result_data['interval_pos_cnt'] = result_data['interval_pos_cnt'].fillna(0)

    #numbers of interval sample
    result_data['interval_total_cnt'] = np.array(tmp4_1.groupby([col_name])['label'].count())

    #ratio of interval
    result_data['interval_percent'] = result_data['interval_total_cnt']/sum(result_data['interval_total_cnt'])

    #Proportion of interval positive samples in all positive samples
    result_data['interval_pos_percent'] = result_data['interval_pos_cnt']/sum(result_data['interval_pos_cnt'])

    #Proportion of interval negative samples in all positive samples
    result_data['interval_neg_percent'] = result_data['interval_neg_cnt']/sum(result_data['interval_neg_cnt'])

    #Proportion of positive samples in each interval
    result_data['pos_percent'] = list(np.array(result_data['interval_pos_cnt'])/np.array(result_data['interval_total_cnt']))
    #Proportion of negative samples in each interval
    result_data['neg_percent'] = list(np.array(result_data['interval_neg_cnt'])/np.array(result_data['interval_total_cnt']))

    #calculation of WOE and IV
    result_data['pyi'] = result_data['interval_pos_cnt']/sum(result_data['interval_pos_cnt'])
    result_data['pni'] = result_data['interval_neg_cnt']/sum(result_data['interval_neg_cnt'])
    diff = result_data['pyi']-result_data['pni']
    divide = result_data['pyi']/result_data['pni']
    result_data['woe'] = [math.log(x) if x!=0 else math.log(x+0.00000001) for x in divide]
    result_data['woe'] = result_data['woe'].fillna(0)

    result_data['IV'] = list(np.array(diff)*np.array(result_data['woe']))
    result_data['IV'] = result_data['IV'].fillna(0)
    result_data['IV_sum'] = sum(result_data['IV'])

    del result_data['pyi']
    del result_data['pni']
    return result_data
```

```python
def get_iv(df):
    tmp_result = pd.DataFrame()
    kk11=list(df.columns)
    for col in kk11:
        if col!= 'label':
            tmpdf = df[[col,'label']]
            tmp_result = tmp_result.append(discrete_iv(tmpdf,col))
    return tmp_result
```
